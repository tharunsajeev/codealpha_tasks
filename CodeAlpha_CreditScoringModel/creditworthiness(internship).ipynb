```python
df=df.drop(columns=['ID','Purpose'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df['Housing']=le.fit_transform(df['Housing'])
df['Saving accounts']=le.fit_transform(df['Saving accounts'])
df['Checking account']=le.fit_transform(df['Checking account'])
```

```python
df.Risk.unique()
```

```python
df['Risk']=le.fit_transform(df['Risk'])
```

```python
df.head(2)
```

```python
X=df.drop(columns=['Risk'])
X
```

```python
y=df['Risk']
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
x_scaled=s.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'max_depth':list(range(1,10)),'n_jobs':[1,2],'random_state':list(range(1,10))}
```

```python
clf=GridSearchCV(RandomForestClassifier(),params,cv=2,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
print('best estimators are:',clf.best_estimator_)
print('best paramters are:',clf.best_params_)
print('best score:',clf.best_score_)
```

```python
model=RandomForestClassifier(max_depth=7,n_jobs=1,random_state=5)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
